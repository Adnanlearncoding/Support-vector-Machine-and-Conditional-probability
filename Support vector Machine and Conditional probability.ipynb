```python
# to covert our dictionary data into data frame 
df=pd.DataFrame(iris.data,columns = iris.feature_names)
```

```python
#now our data  is in datadrame
df
```

```python
# to add target column
df['target'] = iris.target
```

```python
df
```

```python
df.target
```

```python
# if we want to see only 1 and like this you check 0 and 2
df[df.target == 1].head()
```

```python
# TO add new column
df['flower_name'] = df.target.apply(lambda x :iris.target_names[x])
df.head()
```

```python
#lets creat 3 differnt datasets
df1=df[:50] # from 0 index to 50
df2=df[50:100] # from 50 to 100
df3=df[100:] # 100 to remaining
```

```python
# creating a graph btween sepal
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'], color='r',marker='*')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'], color='g',marker='+')
plt.xlabel('sepal length cm')
plt.ylabel('sepal width cm')
```

```python
# creating a graph btween petal like above
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], color='r',marker='*')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'], color='g',marker='+')
plt.xlabel('petal length cm')
plt.ylabel('petal width cm')
```

```python
#above in first graph we draw line linear or huper line while
# 2nd graph, we will draw to seprate line
```

```python
from sklearn.model_selection import train_test_split
```

```python
# drop the dependent variable
x = df.drop(['target','flower_name'], axis='columns')
x.head()
```

```python
# in y we need dependent variable
y = df.target
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
# to creat svm model
from sklearn.svm import SVC
svm_model = SVC()
```

```python
svm_model.fit(x_train,y_train)
```

```python
svm_model.score(x_test,y_test)
```

```python
#sample of x_test to predict some values
x_test
```

```python
svm_model.predict([[5.9,3.2,4.8,1.8]])
```

```python
# above shows if we have features value these. then it shows target 2 which santos
```

```python
#lets  change the value of c
model1=SVC(C=10,gamma=5)
model1.fit(x_train,y_train)
model1.score(x_test,y_test)
```

```python
# if our data is mixed then we will use kernal function
model_linear=SVC(kernel='linear')
model_linear.fit(x_train,y_train)
model_linear.score(x_test,y_test)
```
